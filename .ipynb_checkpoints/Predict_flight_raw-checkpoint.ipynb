```python
import pandas as pd
train=pd.read_excel('Data_Train.xlsx')
train.head()
```

```python
test=pd.read_excel('Test_set.xlsx')
test.head()
```

```python
sub=pd.read_excel('Sample_submission.xlsx')
sub.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train.dropna(how='any',axis=0,inplace=True)
```

```python
train['Date_Journey']=pd.to_datetime(train.Date_of_Journey)
test['Date_Journey']=pd.to_datetime(test.Date_of_Journey)
```

```python
train.head()
```

```python
from datetime import datetime, date
```

```python
train['month'] = train.Date_of_Journey.apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%m'))
train['year'] = train.Date_of_Journey.apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%Y'))
train['day']=train.Date_of_Journey.apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%d'))
```

```python
test['month'] = test.Date_of_Journey.apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%m'))
test['year'] = test.Date_of_Journey.apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%Y'))
test['day']=test.Date_of_Journey.apply(lambda x: datetime.strptime(x, '%d/%m/%Y').strftime('%d'))
```

```python
train['Weekday']=train.Date_Journey.dt.weekday_name
test['Weekday']=test.Date_Journey.dt.weekday_name
```

```python
train.drop(['Date_of_Journey','day','year','Date_Journey'],axis=1,inplace=True)
test.drop(['Date_of_Journey','day','year','Date_Journey'],axis=1,inplace=True)
```

```python
train.head(10)
```

```python
train['Dep_hour'] = train.Dep_Time.apply(lambda x: datetime.strptime(x, '%H:%M').strftime('%H'))
train['Dep_min'] = train.Dep_Time.apply(lambda x: datetime.strptime(x, '%H:%M').strftime('%M'))
test['Dep_hour'] = test.Dep_Time.apply(lambda x: datetime.strptime(x, '%H:%M').strftime('%H'))
test['Dep_min'] = test.Dep_Time.apply(lambda x: datetime.strptime(x, '%H:%M').strftime('%M'))
```

```python
train['Arrival_hour']=train['Arrival_Time'].str.split(':').str[0]
train['Arrival_min']=train['Arrival_Time'].str.split(':').str[1]
test['Arrival_hour']=test['Arrival_Time'].str.split(':').str[0]
test['Arrival_min']=test['Arrival_Time'].str.split(':').str[1]
```

```python
train['Dur_hour']=train['Duration'].str.split(' ').str[0]
train['Dur_min']=train['Duration'].str.split(' ').str[1]
test['Dur_hour']=test['Duration'].str.split(' ').str[0]
test['Dur_min']=test['Duration'].str.split(' ').str[1]
```

```python
train['Arrival_min2']=train['Arrival_min'].str.split(' ').str[0]
train['Arrival_min1']=train['Arrival_min'].str.split(' ').str[1]
test['Arrival_min2']=test['Arrival_min'].str.split(' ').str[0]
test['Arrival_min1']=test['Arrival_min'].str.split(' ').str[1]
```

```python
test
```

```python
train.drop(['Arrival_min','Arrival_min1'],axis=1,inplace=True)
test.drop(['Arrival_min','Arrival_min1'],axis=1,inplace=True)
```

```python
train.dtypes
```

```python
train['Stop1']=train['Route'].str.split('→').str[0]
train['Stop2']=train['Route'].str.split('→').str[1]
train['Stop3']=train['Route'].str.split('→').str[2]
train['Stop4']=train['Route'].str.split('→').str[3]

test['Stop1']=test['Route'].str.split('→').str[0]
test['Stop2']=test['Route'].str.split('→').str[1]
test['Stop3']=test['Route'].str.split('→').str[2]
test['Stop4']=test['Route'].str.split('→').str[3]
```

```python
train
```

```python
train.Dur_min.fillna('0m',inplace=True)
test.Dur_min.fillna('0m',inplace=True)
```

```python
train['Dur_hour']=train['Dur_hour'].map(lambda x:x.rstrip('h'))
test['Dur_hour']=test['Dur_hour'].map(lambda x:x.rstrip('h'))
```

```python
train.Dur_hour.value_counts()
```

```python
train.loc[train['Dur_hour'] == '5m']
```

```python
train.drop(train[train.Dur_hour=='5m'].index, inplace=True)
```

```python
test.loc[test['Dur_hour'] == '5m']
```

```python
test.drop(test[test.Dur_hour=='5m'].index, inplace=True)
```

```python
test.Dur_hour.value_counts()
```

```python
train['Dur_min'] = train['Dur_min'].astype(str).str.replace('m', '')
test['Dur_min'] = test['Dur_min'].astype(str).str.replace('m', '')
```

```python
train.drop(['Route','Dep_Time','Arrival_Time','Duration'],axis=1,inplace=True)
test.drop(['Route','Dep_Time','Arrival_Time','Duration'],axis=1,inplace=True)
```

```python
train.Stop1.fillna('0',inplace=True)
train.Stop2.fillna('0',inplace=True)
train.Stop3.fillna('0',inplace=True)
train.Stop4.fillna('0',inplace=True)

test.Stop1.fillna('0',inplace=True)
test.Stop2.fillna('0',inplace=True)
test.Stop3.fillna('0',inplace=True)
test.Stop4.fillna('0',inplace=True)
```

```python
train['Total_stops'] = train['Total_Stops'].astype(str).str.replace('stop', '')
test['Total_stops'] = test['Total_Stops'].astype(str).str.replace('stop', '')
```

```python
train['Total_stops1'] = train['Total_stops'].astype(str).str.replace('s', '')
test['Total_stops1'] = test['Total_stops'].astype(str).str.replace('s', '')
```

```python
train.drop(['Total_Stops','Total_stops'],axis=1,inplace=True)
test.drop(['Total_Stops','Total_stops'],axis=1,inplace=True)
```

```python
train.Total_stops1.replace('non-',0,inplace=True)
```
